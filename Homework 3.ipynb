### 1. Для чего и в каких случаях полезны различные варианты усреднения для метрик качества классификации: micro, macro, weighted?


Все эти метрики применяются для задач классификации

**Macro** — вычисление метрики для каждого класса и получение невзвешенного среднего

В случае с макро-вариантом усреднение проводится для
величин, которые уже не чувствительны к соотношению размеров классов, и поэтому
каждый класс внесет равный вклад в итоговую метрику.

**Micro** — при микро-усреднении сначала характеристики (TP, FP, и т.д.) усредняются по всем классам, а
затем вычисляется итоговая двухклассовая метрика — например, точность, полнота или F-мера.

Если классы отличаются по мощности, то при микро-усреднении они практически никак не будут влиять на результат, поскольку их вклад в средние TP, FP, FN и
TN будет незначителен. 

**Weighted** — вычисление метрик для каждого класса и получение взвешенного среднего по числу выборок на каждый класс.

### 2. В чём разница между моделями xgboost, lightgbm и catboost или какие их основные особенности?

Самыми заметными отличиями этих алгоритмов является их подход к категориальным признакам и скорость обучения.

LighGBM - самый быстрый, а CatBoost - медленный.

CatBoost и LighGBM могут работать с категориальными признаками.LighGBM может работать только с числовыми значениями в категориальных признаках, а CatBoost еще и со строковыми. 

CatBoost и LighGBM могут заполнять пропущенные значения.

XGBoost отличается от алгоритмов CatBoost и LighGBM. XGBoost не может обрабатывать категориальные функции, он принимает только числовые данные. 

В LighGBM, категориальные переменные функции необходимо преобразовать в целочисленные переменные.Этот алгоритм не позволяет передавать строковые данные в параметры категориальных переменных.

У CatBoost есть встроенные широкие возможности визуализации.
