1. В чём различие между зависимыми и независимыми выборками?

* В зависимых выборках каждому случаю из одной выборки соответствует один случай из второй выборки.
* В независимых выборках, такая взаимосвязь отсутствует.

2. Когда применяются параметрические статистические критерии, а когда — их непараметрические аналоги?

* Параметрические критерии, в отличии от непараметрических, учитывают параметры распределения к которому применяются. 
* Параметрические критерии могут оказаться несколько более мощными, чем непараметрические.
* Если используемый критерий не выявил различия - следует применить более мощный, но одновременно и более трудоемкий критерий.
